```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
tips = sns.load_dataset('tips')
```

```python

# Default catplot (strip plot)
sns.catplot(data=tips, x="day", y="total_bill", kind="strip")
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box", hue="sex")
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="violin", split=True)
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="violin", hue="sex", split=True)
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="bar")
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="bar", hue="sex")
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="bar", hue="sex", col="time")
plt.show()
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="bar", hue="sex", col="time", row="smoker")
plt.show()
```
